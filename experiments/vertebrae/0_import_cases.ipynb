```python
import json
import nibabel as nib
import numpy as np

data_json_path = "/nfs/data/nii/data1/Analysis/zanderch___HU_Messung/ANALYSIS_seg/data.json"
with open(data_json_path) as f:
    data = json.load(f)

mask_paths = data["val"][0]   # list of [path]
img_paths  = data["val"][1]   # list of [path, ...] — 1 or 2 per subject
img_meta   = data["json"][1]  # list of {pid#sid: [{...}]}

issues = []
for i, (mask_list, imgs_list) in enumerate(zip(mask_paths, img_paths)):
    subject_key = list(img_meta[i].keys())[0] if img_meta[i] else f"subject_{i}"
    mask_img   = nib.load(mask_list[0])
    mask_shape = mask_img.shape
    mask_zooms = np.array(mask_img.header.get_zooms())

    for img_path in imgs_list:
        img        = nib.load(img_path)
        img_shape  = img.shape
        img_zooms  = np.array(img.header.get_zooms())
        same_shape = img_shape == mask_shape
        same_zooms = np.allclose(img_zooms, mask_zooms, atol=1e-3)
        if not (same_shape and same_zooms):
            issues.append({
                "subject":    subject_key,
                "img":        img_path.split("/")[-1],
                "img_shape":  img_shape,  "mask_shape":  mask_shape,
                "img_zooms":  img_zooms.round(3).tolist(),
                "mask_zooms": mask_zooms.round(3).tolist(),
                "same_shape": same_shape, "same_zooms":  same_zooms,
            })

if issues:
    print(f"{len(issues)} geometry mismatch(es):\n")
    for r in issues:
        print(f"  {r['subject']}  →  {r['img']}")
        if not r["same_shape"]:
            print(f"    shape   img={r['img_shape']}  mask={r['mask_shape']}")
        if not r["same_zooms"]:
            print(f"    zooms   img={r['img_zooms']}  mask={r['mask_zooms']}")
else:
    print("All images match their mask geometry.")
```

```python
subject_pairs = {}  # key: "pid#sid"  →  (img_path, mask_path)

for i, (mask_list, imgs_list) in enumerate(zip(mask_paths, img_paths)):
    subject_key = list(img_meta[i].keys())[0] if img_meta[i] else f"subject_{i}"
    mask_path   = mask_list[0]
    mask_img    = nib.load(mask_path)
    mask_shape  = mask_img.shape
    mask_zooms  = np.array(mask_img.header.get_zooms())

    matched = None
    for img_path in imgs_list:
        img = nib.load(img_path)
        if img.shape == mask_shape and np.allclose(np.array(img.header.get_zooms()), mask_zooms, atol=1e-3):
            matched = img_path
            break

    if matched is None:
        print(f"ERROR: no matching image for {subject_key}  (mask: {mask_path})")
    else:
        subject_pairs[subject_key] = (matched, mask_path)

print(f"{len(subject_pairs)} subjects with matched (img, mask) pairs")
subject_pairs
```

```python
all_ok = True
for subject_key, (img_path, mask_path) in subject_pairs.items():
    img  = nib.load(img_path)
    mask = nib.load(mask_path)
    same_shape = img.shape == mask.shape
    same_zooms = np.allclose(img.header.get_zooms(), mask.header.get_zooms(), atol=1e-3)
    same_affine = np.allclose(img.affine, mask.affine, atol=1e-3)
    if not (same_shape and same_zooms and same_affine):
        all_ok = False
        print(f"MISMATCH  {subject_key}")
        if not same_shape:
            print(f"  shape   img={img.shape}  mask={mask.shape}")
        if not same_zooms:
            print(f"  zooms   img={np.array(img.header.get_zooms()).round(3)}  mask={np.array(mask.header.get_zooms()).round(3)}")
        if not same_affine:
            print(f"  affine differs")

if all_ok:
    print(f"All {len(subject_pairs)} pairs have matching geometry (shape, zooms, affine).")
```

```python
import shutil

images_dir = task_dir + "/imagesTr"
labels_dir = task_dir + "/labelsTr"
os.makedirs(images_dir, exist_ok=True)
os.makedirs(labels_dir, exist_ok=True)

for subject_key, (img_path, mask_path) in subject_pairs.items():
    case_id = subject_key.replace("#", "_")
    dst_img  = f"{images_dir}/{case_id}_0000.nii.gz"
    dst_mask = f"{labels_dir}/{case_id}.nii.gz"

    # image: load and re-save as .nii.gz (handles .nii → .nii.gz compression)
    nib.save(nib.load(img_path), dst_img)

    # mask: already .nii.gz, plain copy is enough
    shutil.copy2(mask_path, dst_mask)

    print(f"  {case_id}")

print(f"\nDone: {len(subject_pairs)} cases written to {task_dir}")
```
